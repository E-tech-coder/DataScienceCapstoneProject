**Exploration for Test Data Set**

```python
df = pd.read_csv("linkedin_experience_annotated.csv")
df.head(20)
```

```python
print("Anzahl Positionen:", len(df))
print("Anzahl Personen (CVs):", df["person_id"].nunique())
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
df["status"].value_counts()
```

```python
active_per_person = (
    df[df["status"] == "ACTIVE"]
    .groupby("person_id")
    .size()
)

active_per_person.value_counts().sort_index()
```

```python
(df[df["status"] == "ACTIVE"]["seniority"].value_counts())
```

```python
df.groupby("seniority")["position"].apply(lambda x: x.sample(min(5, len(x))))
```

```python
df["department"].value_counts()
```

```python
(df[df["status"] == "ACTIVE"]["department"].value_counts())
```

```python
df.groupby("department")["position"].apply(lambda x: x.sample(min(5, len(x))))
```
