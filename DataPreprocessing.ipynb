```python
import pandas as pd
```

```python
pd.read_csv("/content/DataScienceCapstoneProject/linkedin_experience_annotated.csv")
```

```python
import json

with open('/content/DataScienceCapstoneProject/linkedin-cvs-not-annotated.json',
    "r", encoding = "utf-8"
) as f :
  data = json.load(f)
```

```python
row = []
for person_id, cv in enumerate(data):
  for job in cv :
    job_row = job
    job_row["person_id"] = person_id
    row.append(job_row)
```

```python
df_noLabel = pd.DataFrame(row)
```

```python
from numpy._core.defchararray import index
df_noLabel.to_csv("/content/DataScienceCapstoneProject/linkedin-cvs-not-annotated.csv", index=False)
```
